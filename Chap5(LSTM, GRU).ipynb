### LSTM(Long Short-Term Memory)
1. 기존의 순환 신경망 모델의 한계점을 보완 
2. 장기기억을 담당하는 부분을 추가 
3. 작동원리
    - C는 장기기억을 담당하는 부분으로 곱하기 부분은 기존 정보를 얼마나 남길 것인지 더하기 부분은 현 데이터와 기존 은닉 데이터의 정보를 추가 
    - 망각 게이트로써 기존 정보들로 구성되어 있는 셀 상태의 값을 얼마나 잊어버릴것인지 정하는 부분 
    - 입력 게이트는 어떤 정보를 얼만큼 셀에 새롭게 저장할 것인지 정하는 부분 
    - 정리 
        - 현 시점의 새로운 입력값과 직전 시점의 은닉 상태 값의 조합으로 기존의 셀 상태의 정보를 얼만큼 전달할지 정하고 어떤 정보를 얼만큼의 비중으로 더할지도 정하는 것 


### GRU(Gated Recurrent Unit)
1. LSTM보다 간단한 구조이면서 성능 면에서 밀리지 않는 RNN의 변형 형태 
2. GRU는 LSTM과 달리 셀 상태와 은닉 상태를 분리하지 않고 은닉 상태 하나로 합침 
3. 업데이트 게이트는 현재 시점의 새로운 input과 직전 시점의 은닉 상태 값에 가중치를 곱해 시그모이드 함수를 통과시켜 업데이트 비중을 결정 
4. 리셋 게이트는 업데이트 게이트와 같은 입력을 받아 동일하게 시그모이드 함수를 통해 비중을 정함 
5. 기존 은닉 상태에 가중치가 곱해진 값과 새로운 입력값을 입력으로 받아 가중치를 곱한 후 하이퍼볼릭 탄젠트 함수를 통과해 새로운 정보의 값을 리턴 
6. 새로운 은닉 상태를 구하는 것

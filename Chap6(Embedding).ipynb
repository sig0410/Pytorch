### Embedding
1. one-hot vector로 바꿔서 연산을 하는 것은 의미적 연산과 확장성 측면에서 한계가 있고 one-hot vector간 내적은 항상 0이 나오기 때문에 문장 간 의미적 차이나 유사도를 구하는 것이 불가능 
2. one-hot의 한계를 극복하기 위한 방법 
    - 알파벳이나 단어 같은 기본 단위 요소들을 일정한 길이를 가지는 벡터 공간에 투영 
    - 즉 단어를 벡터화한다고 생각 
3. Word2Vec
    - 임베딩의 대표적인 기법이며 CBOW와 skip-gram이 있음 
    - CBOW 
        - 주변 단어들로부터 가운데 들어갈 단어가 나오도록 임베딩 
        - 문장을 주고 모델을 학습하면 주위 단어들과 가운데 들어갈 단어와의 관계가 학습이 되며 의미적으로 임베딩이 됨
    - skip-gram
        - 중심 단어로 주변 단어들이 나오도록 학습하여 임베딩 벡터를 얻는 방식 
        - CBOW와 반대라고 생각

### 선형회귀분석 모델 
   - w, b를 업데이트하는 전체 과정 

```python
import torch 
import torch.nn as nn
# torch.nn : 신경만 모델들이 포함 
import torch.optim as optim
# torch.optim : 경사하강법 알고리즘 
import torch.nn.init as init 
# torch.nn.init : 텐서에 초깃값을 주기 위해 필요한 함수 
```

```python
num_data = 1000
num_epoch = 500

x = init.uniform_(torch.Tensor(num_data, 1), -10, 10)
# -10 ~ 10 사이의 값으로 [num_data, 1]이라는 텐서를 균등하게 초기화 
noise = init.normal_(torch.FloatTensor(num_data, 1), std = 1)
# 평균이 0 표준편차가 1인 값들을 넣어줌 
y = 2*x + 3
y_noise = y + noise 
# y_noise = y값에 noise를 더해서 실제 분석에서 사용하는 데이터와 유사하게 만듬 
```

```python
print(x)
```

```python
print(noise)
```

```python
model = nn.Linear(1, 1)
# nn.Linear() : 선형회귀 모델을 만드는 함수 (특성의 수, 결과로 나오는 특성의 수, 편차 사용 여부)
# nn.Linear(1, 1) : x랑 y가 하나씩 있기때문에 이렇게 지정 
loss_func = nn.L1Loss()
# L1Loss : L1Loss손실을 뜻함 
```

```python
optimizer = optim.SGD(model.parameters(), lr = 0.01)
# 최적화 함수라고도 하고 경사하강법을 적용하여 오차를 줄이고 최적의 가중치와 편차를 근사할 수 있음
# model.parameters()를 사용하여 선형회귀 모델의 변수 w와 b를 전달
```

```python
label = y_noise 

for i in range(num_epoch):
    optimizer.zero_grad()
    # zero_grad : 각 반복 시 기울기를 0으로 초기화시켜주는 함수 
    # 기울기를 초기화해야 새로운 가중치와 편차에 대해서 새로운 기울기를 구할 수 있음 
    output = model(x)
    # output : model에 x를 넣어서 나온 결과 
    
    loss = loss_func(output, label)
    loss.backward()
    # loss.backward() : 각 변수, w와 b에 대한 기울기가 계산 
    optimizer.step()
    # optimizer.step() : 함수를 호출하여 model.parameters에서 리턴되는 변수들의 기울기에 학습률 0.01을 곱하여 빼줌으로 업데이트 
    
    if i % 10 == 0:
        print(loss.data)
        param_list = list(model.parameters())
        print(param_list[0].item(), param_list[1].item())

# 결과 : y = 2x + 3과 같이 w와 b가 2와 3에 근접한 것을 알 수 있다.
```
